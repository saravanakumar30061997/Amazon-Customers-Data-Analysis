# Question3
Is there any difference between behaviour of frequent viewers and not frequent viewers?

```python
df
```

```python
df.columns
```

```python
df["UserId"].value_counts()
```

```python
df.head()
```

```python
x=df["UserId"].value_counts()
```

```python
def freq(Id):
    if x[Id]>50:
        return "Frequent"
    else:
        return "Not Frequent"
```

```python
freq("A3SGXH7AUHU8GW")
```

```python
df["Frequency"]=df["UserId"].apply(freq)
```

```python
df["Frequency"].value_counts()
```

```python
not_frequent_df=df[df["Frequency"]=="Not Frequent"]
```

```python
frequent_df=df[df["Frequency"]=="Frequent"]
```

```python
f=frequent_df["Score"].value_counts()/len(frequent_df)*100
```

```python
f
```

```python
f.plot(kind="bar")
```

```python
nf=not_frequent_df["Score"].value_counts()/len(not_frequent_df)*100
nf
```

```python
nf.plot(kind="bar")
```

```python
from textblob import TextBlob
```

```python
df
```

```python
TextBlob(df["Summary"][0]).sentiment.polarity
```

```python
df.shape
```

```python
sample=df[:50000]
```

```python
sample
```

```python
polarity=[]


for text in sample["Summary"]:
    try:
        polarity.append(TextBlob(text).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
polarity
```

```python
sample["polarity"]=polarity
```

```python
sample
```

```python
sample_neg=sample[sample["polarity"]<0]
```

```python
sample_pos=sample[sample["polarity"]>0]
```

```python
sample_neg["Summary"]
```

```python
from collections import Counter
```

```python
Counter(sample_neg["Summary"]).most_common()
```

```python
Counter(sample_pos["Summary"]).most_common()
```
